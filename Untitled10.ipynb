```python
import numpy as np
x = np.array([4, 9, 14])
w = np.array([[0.25, 0.35],
              [0.45, 0.55],
              [0.65, 0.75]])
b = np.array([0.1, 0.25])

z = np.dot(x, w) + b
print("output:", z)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

a = sigmoid(z)
print("activated output:", a)

w2 = np.array([[0.1, 0.18]])
b2 = np.array([0.15])

z2 = np.dot(a, w2.T) + b2
a2 = sigmoid(z2)
print("final output:", a2)

y = 0.7
loss = (y - a2) ** 2
print("loss:", loss)

lr = 0.005

dloss_da2 = -2 * (y - a2)
da2_dz2 = a2 * (1 - a2)
dz2_dw2 = a

grad_w2 = dloss_da2 * da2_dz2 * dz2_dw2
w2 = w2 - lr * grad_w2.reshape(2, 1)

print("updated w2:", w2)
```

```python
import tensorflow as tf

x = tf.constant([[8., 12., 18.]])
y = tf.constant([[0.7]])

w1 = tf.constant([
    [0.15, 0.25],
    [0.35, 0.45],
    [0.55, 0.65]
], dtype=tf.float32)

b1 = tf.constant([0.8, 1.0], dtype=tf.float32)

w2 = tf.constant([
    [0.12],
    [0.22]
], dtype=tf.float32)

b2 = tf.constant([0.4], dtype=tf.float32)

z1 = tf.matmul(x, w1) + b1
a1 = tf.sigmoid(z1)

z2 = tf.matmul(a1, w2) + b2
a2 = tf.sigmoid(z2)

loss = tf.square(y - a2)

print("Final Output:", a2.numpy())
print("Loss:", loss.numpy())
```

```python
import torch
import torch.nn as nn

x = torch.tensor([[3., 6., 9.]])
y = torch.tensor([[0.9]])

class FeedForwardNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(3, 2)
        self.fc2 = nn.Linear(2, 1)

        self.fc1.weight.data = torch.tensor(
            [[0.15, 0.35, 0.55],
             [0.25, 0.45, 0.65]]
        )
        self.fc1.bias.data = torch.tensor([0.4, 0.7])

        self.fc2.weight.data = torch.tensor([[0.12, 0.22]])
        self.fc2.bias.data = torch.tensor([0.25])

    def forward(self, x):
        x = torch.sigmoid(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x

model = FeedForwardNN()

output = model(x)
loss = (y - output) ** 2

print("Final Output:", output.item())
print("Loss:", loss.item())
```

```python
import numpy as np

class FeedForwardNN:
    def __init__(self, lr=0.05):
        self.w1 = np.array(
            [[0.15, 0.25],
             [0.35, 0.45],
             [0.55, 0.65]]
        )
        self.b1 = np.array([[0.4, 0.6]])

        self.w2 = np.array([[0.12],
                            [0.22]])
        self.b2 = np.array([[0.25]])

        self.lr = lr

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def forward(self, X):
        self.z1 = np.dot(X, self.w1) + self.b1
        self.a1 = self.sigmoid(self.z1)

        self.z2 = np.dot(self.a1, self.w2) + self.b2
        self.a2 = self.sigmoid(self.z2)
        return self.a2

    def backward(self, X, y):
        d_loss_da2 = self.a2 - y
        d_loss_dz2 = d_loss_da2 * self.sigmoid_derivative(self.a2)

        self.dw2 = np.dot(self.a1.T, d_loss_dz2)
        self.db2 = np.sum(d_loss_dz2, axis=0, keepdims=True)

        d_loss_da1 = np.dot(d_loss_dz2, self.w2.T)
        d_loss_dz1 = d_loss_da1 * self.sigmoid_derivative(self.a1)

        self.dw1 = np.dot(X.T, d_loss_dz1)
        self.db1 = np.sum(d_loss_dz1, axis=0, keepdims=True)

    def update(self):
        self.w1 -= self.lr * self.dw1
        self.b1 -= self.lr * self.db1
        self.w2 -= self.lr * self.dw2
        self.b2 -= self.lr * self.db2

    def train(self, X, y, epochs=800):
        for epoch in range(epochs):
            self.forward(X)
            self.backward(X, y)
            self.update()
            if (epoch + 1) % 100 == 0:
                loss = np.mean((y - self.a2) ** 2)
                print(f"Epoch {epoch+1}, Loss: {loss}")

X = np.array([[2, 3, 4]])
y = np.array([[0.9]])

model = FeedForwardNN(lr=0.05)
model.train(X, y, epochs=800)

print("Final Output:")
print(model.forward(X))
```
